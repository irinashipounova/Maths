```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
# задание 4 - 1ый вариант решения
x = np.linspace(100, 500, 201)
a = np.random.uniform(0.01, 0.02)
plt.plot(x, np.sin(a*x), 'blue')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show
```

```python
# задание 4 - 2ой вариант решения
x = np.linspace(100, 500, 201)
a1 = 0.01
a2 = 0.02
plt.plot(x, np.sin(a1*x), 'blue')
plt.plot(x, np.sin(a2*x), 'green')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show
# видимо, ответ находится между двумя точками пересечения, но как его определить численно, я не знаю.
```
